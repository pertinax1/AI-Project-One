```python
test_df = pd.read_csv("personal_resources/account_predict_data.csv")
display(test_df.head())
```

```python
gender_churn_ratio_df = test_df[["gender", "churn"]]
display(gender_churn_ratio_df.head())
```

```python
column_name = "gender"
```

```python
def determine_churn_ratio(df):
    churn_column = df["churn"]
    return churn_column.sum()/churn_column.count()


churn_sum_df = gender_churn_ratio_df.groupby(by=gender_churn_ratio_df[column_name]).sum()
churn_count_df = gender_churn_ratio_df.groupby(by=gender_churn_ratio_df[column_name]).count()
churn_ratio_df = churn_sum_df/churn_count_df

display(churn_sum_df.head())
display(churn_count_df.head())
display(churn_ratio_df.head())
```
